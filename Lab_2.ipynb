###Kurtosis

Kurtosis measures the shape of the distribution and whether data is heavy-tailed or light-tailed compared to a normal distribution. In simple words, data with high kurtosis tend to have heavy tails, or outliers and data with low kurtosis tend to have light tails, or lack of outliers.

**Leptokurtic (Kurtosis > 3):** Very long and thick tails meaning a higher chance of outliers. Has a sharp peak.

**Platykurtic (Kurtosis < 3):** Thin tail and stretched around the center means most data points are present in high proximity to the mean. Has a flatter peak.

**Mesokurtic (Kurtosis = 3):** Distribution is normal.

###Skewness

The skewness is a measure of how symmetrical the data distribution is. When data deviates from a perfectly symmetrical normal distribution (bell shaped curve), skewness measures the degree of asymmetry of the data.


> **Symmetrical Data:** Skewness = ~0 (between -0.5 and 0.5)

> **Positively Skewed Data:** Skewness > 1

> **Negatively Skewed Data:** Skewness < -1

##Missing Data

###Types of Missing Data

**Missing Completely At Random (MCAR)**

In MCAR, the probability of data being missing is the same for all the observations and there is no relationship between the missing data and any other values observed or unobserved. The missing values are completely independent of other data and there is no pattern.

For example, data could be missing due to human error, some system/equipment failure, loss of sample, etc. For example, suppose we have a dataset of overdue library books. Some data related to the books is missing. This could be because of a human error like the librarian forgetting to enter the data.

**Missing At Random (MAR)**

In MAR, the probability of data being missing is not the same for all observations since there is a relationship between the missing data and other variables in the dataset and there is a pattern to it.

For example, in a dataset of overdue library books, observations with large values of "*Number of overdue books*" have missing values for "*Respondent name*". We can assume that respondents with a large number of overdue books wish to keep their identity anonymous and thus have not answered.

**Missing Not At Random (MNAR)**

MNAR is considered to be the most difficult scenario among the three types of missing data. It is applied when neither MAR nor MCAR apply. In this situation, the probability of being missing is completely different for different values of the same variable. It can occur due to the reluctance of people to provide the required information. A specific group of respondents may not answer some questions in a survey.

For example, in a dataset of overdue library books, observations with "*Books currently checked out*" are missing. We can assume that the individual respondents did not want to disclose what books they had checked out for personal reasons.

###Handling Missing Values

**Removing the columns having missing values** â€” If you have a column with more than 80% missing values, then it is better to drop the column.

**Removing the rows having missing values** â€” If you have a small percentage of rows with missing values, then you can just drop those rows. If say you have 1 Million records, and out of that just 20 rows have missing values, then you can drop those rows since they wonâ€™t take away a lot of information from your data.

**Imputing the missing values** â€” This is usually the preferred way of handling the missing values, especially where the missing values are not more than 80%. Imputing means your fill the missing values with a valid value ,for instance replacing missing values with the average, mode, or median.

```python
#Removing a column

dfFixNulls = df.copy()
dfFixNulls = dfFixNulls.drop(['Cabin'], axis=1)
dfFixNulls.isnull().sum()*100/len(dfFixNulls)
```

```python
#Removing a row
#Delete row with any null values

dfDeleteRows = df.copy()
dfDeleteRows.dropna(axis=0, inplace=True)

print("Original dataset: " + str(len(df)))
print("Dataset with null rows removed: " + str(len(dfDeleteRows)))
```

```python
#or drop a row based on percentage of data missing
#keep rows with minimum 90% of non-nan values

dfDeleteRows = df.copy()
dfDeleteRows = dfDeleteRows.dropna(thresh=0.9*len(dfDeleteRows.columns))

print("Original dataset: " + str(len(df)))
print("Dataset with null rows removed: " + str(len(dfDeleteRows)))
```

```python
#Replace with a specific value
#helps recognize nulls

dfFixNulls['Embarked'] = dfFixNulls['Embarked'].fillna(0)
dfFixNulls['Embarked'].value_counts()
```

```python
#Replace with the mean
#Most common method
#Does not work if the data has outliers

print("Kurtosis of Age is: " + str(dfFixNulls['Age'].kurt()))
print("Skewness of Age is: " + str(dfFixNulls['Age'].skew()))
print("Standard Deviation of Age is: " + str(dfFixNulls['Age'].std()))
#Not too many outliers according to SD, skewness and kurtosis

dfFixNullAccMean = dfFixNulls.copy()
print("Mean: " + str(dfFixNullAccMean['Age'].mean()))

dfFixNullAccMean['Age'] = dfFixNullAccMean['Age'].fillna(dfFixNullAccMean['Age'].mean())
dfFixNullAccMean.isnull().sum()*100/len(dfFixNullAccMean)
```

```python
#Replace by mean according to gender
dfFixNullAccMean = dfFixNulls.copy()

dfFixNullAccMean['Age'] = dfFixNullAccMean['Age'].fillna(dfFixNullAccMean.groupby('Gender')['Age'].transform('mean'))
dfFixNullAccMean.to_excel("dfFixNullAccMean.xlsx", index=False)
dfFixNullAccMean.isnull().sum()*100/len(dfFixNullAccMean)
```

```python
#Replace with the mode
#Mode is the most frequently occurring value
#Mostly used in the case of categorical features

dfFixNullAccMode = dfFixNulls.copy()
print("Mode: " + str(dfFixNullAccMode['Age'].mode()[0]))

dfFixNullAccMode['Gender'] = dfFixNullAccMode['Gender'].fillna(dfFixNullAccMode['Gender'].mode()[0])
dfFixNullAccMode.isnull().sum()*100/len(dfFixNullAccMode)
```

```python
#Replace with median
#The median is the middlemost value
#Useful when data has outliers.
#Assumes the data is MCAR

dfFixNullAccMedian = dfFixNulls.copy()
print("Median: " + str(dfFixNullAccMedian['Age'].median()))

dfFixNullAccMedian['Age']= dfFixNullAccMedian['Age'].fillna(dfFixNullAccMedian['Age'].median())
dfFixNullAccMedian.isnull().sum()*100/len(dfFixNullAccMedian)
```

```python
#Replacing with previous value. This is called forward fill.
#Replacing with the next value. This is called backward fill
#Mostly used in time series data.

dfExample = df.copy()
dfExample['Age'] = dfExample['Age'].fillna(method='bfill')

dfExample = df.copy()
dfExample['Age'] = dfExample['Age'].fillna(method='ffill')
```
